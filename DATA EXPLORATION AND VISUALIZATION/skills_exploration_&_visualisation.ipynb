```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
skills= pd.read_csv('skills.csv')
```

```python
skills
```

```python
skills= skills.drop('Unnamed: 0', axis=1)
skills
```

```python
skills.stack()
```

```python
# color palette

name_colors = ['#78C850','#F08030','#6890F0','#A8B820', '#A8A878','#A040A0', '#F8D030', '#E0C068']
```

```python
# bAR PLOT (count plot)
plt.figure(figsize=(10,6))
sns.countplot(x="name", data=skills, palette= name_colors)
plt.title('Name Against skills')
```
